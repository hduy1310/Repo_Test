```python
class Student:
    #Constructor (hàm khởi tạo), 2 tham số: anme và age
    def __init__(self, name, age):
        #gan gia tri tham so cho thuoc tinh cua doi tuong
        self.name = name
        self.age = age
    def hi(self):
        print(f'My name is {self.name}, I am {self.age} year old.')
```

```python
john = Student('John',20)
paul = Student('Paul',19)
```

```python
john.hi()
```

```python
class Student2:
    #Constructor (hàm khởi tạo), 2 tham số: anme và age
    def __init__(self, name, age):
        #gan gia tri tham so cho thuoc tinh cua doi tuong
        self.name = name
        self.__age = age
    def hi(self):
        print(f'My name is {self.name}, I am {self.__age} year old.')
```

```python
s1 = Student2('John',20)
s2 = Student2('Paul',19)
```
