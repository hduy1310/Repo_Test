```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('D:\\AI Engineer\\DataSample\\USImport.csv',delimiter=';')
df.head()
```

```python
# convert 'Quantity' to integer
df['Quantity'] = df['Quantity'].str.replace(',', '.').astype(float).astype(int)
```

```python
df['Weight'] = df['Weight'].str.replace(',', '.').astype(float)
```

```python
df['Value (US$)'] = df['Value (US$)'].str.replace(',', '.').astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Quantity'],bins=30,kde=True)
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Weight'], bins=30, kde=True)
plt.title('Distribution of Weight')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Value (US$)'], bins=20, kde=True)
plt.title('Distribution of Import Values')
plt.xlabel('Value (US$)')
plt.ylabel('Frequency')
plt.show()
```

```python

    #Trung bnh cng
print(f'Mean of quantity: {df['Quantity'].mean():.2f}')
    #Trung v: Gi tr nm gia khi sp xp t nh n ln
print(f'Median of quantity: {df['Quantity'].median():.2f}')
    #Mode Gi tr xut hin nhiu ln nht trong tp d liu
print(f'Mode of quantity: {df['Quantity'].mode()[0]}')
    #do lech chuan
print(f'Std of quantity: {df['Quantity'].std():.2f}')
    #he so bien thien
print(f'Std of quantity: {(df['Quantity'].std()/df['Quantity'].mean())*100:.2f}%')


print(f'Min of quantity: {df['Quantity'].min():.2f}')
print(f'Max of quantity: {df['Quantity'].max():.2f}')
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['Quantity'])
plt.title('Boxplot of Quantity')
plt.xlabel('Quantity')
plt.show()
```
