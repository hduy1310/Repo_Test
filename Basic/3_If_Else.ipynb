```python
score=float(input('Enter your score: '))
if score >= 90:
    print('Xếp loại xuất sắc, điểm của bạn là: ',score)
elif score >= 75:
    print(f"Xếp loại khá, điểm của bạn là: {score:.3f}")
else:
    print("Xếp loại trung bình")
```

```python
payment = int(input('Enter payment amount: '))
member = input('Are you a member (Y/N)?: ')
if member == 'Y' or member == 'y':
    membership = input('Enter membership type (Vip/Gold/Diamond): ')
    if membership == 'Vip': payment *= 0.9
    elif membership == 'Gold': payment *= 0.85
    elif membership == 'Diamond': payment *= 0.75
    else: print('Invalid membership type')
 
print(f'Final payment amount: {payment:.2f} VND')
 
```

```python
amount = float(input('Enter amount: '))
member = input('Enter Y/N member: ')
if member == 'y' or member == 'Y':
    member_level = input('Enter member ship type (diamond/gold/silver/bronze):')
    if member_level == 'diamond': discount = 0.1
    elif member_level == 'gold': discount = 0.09
    elif member_level == 'silver': discount = 0.08
    elif member_level == 'bronze': discount = 0.07

print(amount*discount)
print(amount-(amount*discount))
```
