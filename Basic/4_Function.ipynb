```python
def print_line(length):
    for _ in range(length):
            print('*',end=' ')
    print()    
```

```python
print_line(5)
```

```python
n=5
for i in range(n):
    print_line(i+1)
```

```python
def print_line2(length,c):
    for _ in range(length):
            print(c,end=' ')
    print()    

print_line2(5,'%')
```

```python
def print_line3(n):
    for i in range(n+1):
        for j in range(i):
                print('*',end=' ')
        print()

print_line3(5)
```

```python
def print_line4(length,c,end_line='\n'):
    for _ in range(length):
            print(c,end=' ')
    print(end_line, end='')    

n=5
for i in range(n):
    print_line4(n-1-i,' ', end_line='')
    print_line4(2*i+1,'*', end_line='\n')

```
