```python
# dictionary is collection of key-value pairs
# keys are unique and immutable
# values can be accessed via keys
my_dict = {'apple': 'quả táo','banana':'quả chuối','orange':'quả cam'}
```

```python
print(my_dict)
```

```python
print(my_dict['apple'])
```

```python
my_dict['mango']='quả xoài'
print(my_dict)
```

```python
my_dict.get['mango'] #edit = insert
```

```python
fruit = input('Enter a fruit in English: ')
if fruit in my_dict:
    print(f'Vietnamese name: {my_dict[fruit]}')
else:
    print('Khong co')
```

```python
# insert new word
new_fruit = input('Enter a new fruit in English: ')
if new_fruit not in my_dict:
    trans = input(f'Enter the Vietnamese translation for {new_fruit}: ')
    my_dict[new_fruit] = trans
    print('Updated dictionary:', my_dict)
else:
    print(f'{new_fruit} already exists in the dictionary.')
```

```python
my_dict['pineapple']='quả dứa'
print(my_dict)
```

```python
del my_dict['pineapple']
print(my_dict)
```

```python
my_dict.pop('mango')
print(my_dict)
```
