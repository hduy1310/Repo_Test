```python
#bài 1
class SinhVien:
    def __init__(self, ten, tuoi):
        self.ten = ten
        self.tuoi = tuoi

    def gioi_thieu(self):
        print(f"Tôi tên là {self.ten}, {self.tuoi} tuổi.")
sv = SinhVien(input("Nhập Tên sinh viên: "), input("tuổi: "))
sv.gioi_thieu()
```

```python
#bài 2 và 3
class HinhChuNhat:
    def __init__(self, chieu_dai, chieu_rong):
        self.chieu_dai = chieu_dai
        self.chieu_rong = chieu_rong

    def dien_tich(self):
        dt = self.chieu_dai * self.chieu_rong
        print(f'diện tích HCN là: {dt}')
    # thêm phương thức vào lớp
    def chu_vi(self):
        cv = (self.chieu_dai + self.chieu_rong) * 2
        print(f'chu vi HCN là: {cv}')

hcn = HinhChuNhat(float(input("chiều dài: ")), float(input("chiều rộng: ")))
hcn.dien_tich()
hcn.chu_vi()
```

```python

# bài 5
class HocSinh:
    def __init__(self, ten, diem):
        self.ten = ten
        self.__diem = diem  # thuộc tính riêng tư

    def get_diem(self):
        return self.__diem

    def set_diem(self, diem_moi):
        if 0 <= diem_moi <= 10:
            self.__diem = diem_moi
        else:
            print("Điểm không hợp lệ!")

hs = HocSinh(input('Tên: '), float(input('điểm: ')))
print(f'Điểm ban đầu: {hs.get_diem()}')
hs.set_diem(float(input('thay đổi điểm: ')))
print(f'Điểm sau khi thay đổi: {hs.get_diem()}')
```

```python
#bài 9
class PhuongTien:
    def __init__(self, loai_phuong_tien):
        self.loai_phuong_tien = loai_phuong_tien

    def di_chuyen(self):
        print(f"{self.loai_phuong_tien} đang di chuyển.")

class XeMay(PhuongTien):
    def __init__(self, loai_phuong_tien, toc_do):
        super().__init__(loai_phuong_tien)
        self.toc_do = toc_do

    def thong_tin(self):
        print(f"{self.loai_phuong_tien} chạy với tốc độ {self.toc_do} km/h.")

class OTo(PhuongTien):
    def __init__(self, loai_phuong_tien, so_ghe):
        super().__init__(loai_phuong_tien)
        self.so_ghe = so_ghe

    def thong_tin(self):
        print(f"{self.loai_phuong_tien} có {self.so_ghe} ghế.")

xe_may = XeMay("Xe máy", input("tốc độ: "))
xe_may.di_chuyen()
xe_may.thong_tin()

oto = OTo("Ô tô", input('số ghế'))
oto.di_chuyen()
oto.thong_tin()
```
