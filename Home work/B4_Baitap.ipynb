```python
#1.Ham tim so lon nhat (a,b,c)
def find_max(a,b,c):
    max = a
    if(b>max):
        max = b
    elif(c>max):
        max = c
    print(f'Maximum is: {max}')

#call function
x=int(input('Enter a:'))
y=int(input('Enter b:'))
z=int(input('Enter c:'))

find_max(x,y,z)
```

```python
#2.Convert Celsius to Fahrenheit. Formular: F = (9/5) * C + 32
def convert_fahrenheit_to_celsius (x):
    result = (9/5) * x + 32
    print(f'{x:.2f} oF = {result:.2f} oC')

#call function
x=int(input('Enter Celsius:'))
convert_fahrenheit_to_celsius(x)
```

```python
#3.In hinh chu nhat voi n,m input from user
def print_rectange(n,m):
    n = int(input('Enter n: '))
    m = int(input('Enter m: '))
    for i in range(n):
        print('* '*m,end="\n")
```

```python
print_rectange(n,m)
```

```python
#4.Viết hàm kiểm tra 1 số là số chính phương hay không
def check(x):
    x = float(input('Enter x: '))
    sqrt_x = ((x**0.5)%1)
    if sqrt_x == 0:
        print(f'{x} là số chính phương.')
    else:
        print(f'{x} không phải là số chính phương.')
```

#5.Viết hàm tính tiền điện
● Dưới 50 kWh: 1.678 VND/kWh
● Từ 51 đến 100 kWh: 1.734 VND/kWh
● Từ 101 đến 200 kWh: 2.014 VND/kWh
● Trên 200 kWh: 2.536 VND/kWh
Viết hàm main sử dụng vòng lặp yêu cầu người dùng nhập số điện tiêu
thụ, gọi hàm tính tiền rồi in ra giá phải trả, sau đó hỏi người dùng có
muốn tiếp tục không, nếu có thì quay lại từ đầu hỏi người dùng nhập
số điện.

```python
#5.Tính tiền điện
def auto_tien_dien():
    while True:
        print_menu()
        qty = input_kwh()
        calculate(qty)

def print_menu():
    print('1.Số Kwh <= 50 đơn giá là: 1.678')
    print('2.Số Kwh >50 đơn giá là: 1.734')
    print('3.Số Kwh >100 đơn giá là: 2.014')
    print('4.Số Kwh >200 đơn giá là: 2.536')

def input_kwh():
    qty = int(input('Nhap sp luong dien (kwh): '))
    if qty > 0:
        return qty
    return qty

def calculate(qty):
    if qty < 0: print('Invalid value, Pls input again!')
    price1 = 1.678
    qty1 = 50

    price2 = 1.734
    qty2 = 100

    price3 = 2.014
    qty3 = 200

    price4 = 2.536
    qty4 = 201

    result = float(0)
    if qty <= qty1:
        result = qty*price1

    elif qty > qty1 and qty <= qty2:
        result = qty1*price1 + (qty-qty1)*price2

    elif qty > qty2 and qty <= qty3:
        result = qty1*price1 + (qty2-qty1)*price2 + (qty-qty2)*price3

    elif qty > qty3:
        result = qty1*price1 + (qty2-qty1)*price2 + (qty3-qty2)*price3 + (qty-qty3)*price4

    print(f'Total amount is: {result}',end="\n")


if __name__ == '__main__':
    auto_tien_dien()
```
