```python
#7.Tính tổng các số dương nhập vào, số âm dừng.
number = int(input('Enter so nguyen: '))
result=0
while(number > 0):    
    result = result+number
    number = int(input('Enter so nguyen: '))
print(result)
```

```python
#8.In số fibonacci <100:
n=int(input('Enter n:'))
i=0
j=1
print(f'0\n1')
while (i+j < n):
    x=i+j
    i=j
    j=x
    print(x,end="\n")
```
