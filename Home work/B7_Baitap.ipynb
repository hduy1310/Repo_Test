```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Tạo một DataFrame chứa thông tin về sản phẩm, danh mục, và doanh thu. 

df_products = pd.DataFrame({
    'product_id': [1, 2, 3, 4, 5,6],
    'category': ['Computer', 'Laptop', 'Tablet', 'Smartphone', 'Smartwatch', 'Laptop'],
    'revenue': [1999, 2999, 1500, 1999, 299, 1]
})

# TODO: Sử dụng groupby() để tính tổng doanh thu theo từng danh mục
```

```python
grouped_category_df = df_products.groupby('category')
result=grouped_category_df.sum()
print(result['revenue'])
```

```python
# Tạo một DataFrame về thông tin nhân viên và lương và phòng ban. 

df_employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Jane', 'Jim', 'Jill', 'Jack'],
    'salary': [50000, 55000, 60000, 65000, 70000],
    'department': ['IT', 'HR', 'IT', 'HR', 'IT']
})

# TODO: Sử dụng groupby() để tính trung bình lương theo từng phòng ban
```

```python
grouped_dep_df = df_employees.groupby('department')
result2 = grouped_dep_df['salary'].mean()
print(result2)
```
