```python
#1.Create class sinh vien and print method gioi_thieu() : ten tuoi
class sinh_vien:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def gioi_thieu(self):
        print(f'My name {self.name}, I am {self.age} year old.')
        
```

```python
Nam = sinh_vien('Nam',27)
My = sinh_vien('My',25)
Nam.gioi_thieu()
My.gioi_thieu()
```

```python
#2.Create class rectangle
class Rectango:
        def __init__(self,width,height):
            self.width = width
            self.height = height
        def dien_tich(self):
            print('Dien tich = ',self.width * self.height)
        def chu_vi(self):
             print('Chu vi = ',self.width*2 + self.height*2)
             
                
```

```python
a = Rectango(5,7)
a.dien_tich()
a.chu_vi()
```
